```python
import numpy as np
import pandas as pd
```

```python
data_plot = pd.DataFrame(np.random.rand(10, 4),index=pd.date_range('1/1/2000', periods=10),columns=['Amazon', 'Apple', 'Meta', 'Alphabet'])
data_plot
```

```python
data_plot.plot()
```

```python
data_bar = pd.DataFrame(np.random.randint(low=2, high=100, size=(10,4)),columns=['Amazon', 'Apple', 'Meta', 'Alphabet'])
data_bar.plot.bar()
```

```python
data_bar.plot.bar(stacked=True)#default value for stacked is false
```

```python
#barh-horizontal
data_bar.plot.barh()
```

```python
data_hist = pd.DataFrame({'Maths':np.random.randint(low=30, high=100, size=100),'Science':np.random.randint(low=10, high=80, size=100),
                          'Language':np.random.randint(low=40, high=70, size=100)},
                         columns=['Maths', 'Science', 'Language'])
data_hist
```

```python
data_hist.plot()
```

```python
data_hist.Maths.plot.hist(bins=50)
```

```python
#kde-density
data_hist.plot.kde()
```

```python
data_pie = pd.DataFrame(3 * np.random.rand(5), index=['Rent', 'Grocery', 'Outing', 'EMI', 'Misc'], columns=['Expenses'])
data_pie.plot.pie(subplots=True)
```

```python
import matplotlib.pyplot as plt
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]

y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]

y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x)
#[<matplotlib.lines.Line2D at 0x7c79a0bf8220>]-top avoid this use plot.show()
```

```python
plt.plot(x,y1)
```

```python
plt.plot(x,y1,label='y1',linewidth=10)
plt.legend()
plt.plot(x,y2,label='y2')
plt.xlabel('x')
plt.xlabel('y')
```
